2. Checking for missing values

Only printing the missing values

```python
null_train = train.isna().sum()
print(null_train[null_train > 0])
```

```python
null_test = test.isna().sum()
print(null_test[null_test > 0])
```

Printing missing values by percentage

```python
null_train = null_train.to_frame(name = 'null_counts')
null_train = null_train[null_train['null_counts'] > 0]
null_train['percentage'] = (null_train['null_counts'] / len(train)) * 100

print(null_train.sort_values(by = 'null_counts', ascending = False))
```

```python
null_test = null_test.to_frame(name='null_counts')
null_test = null_test[null_test['null_counts'] > 0]
null_test['percentage'] = (null_test['null_counts'] / len(test)) * 100

print(null_test.sort_values(ascending=False, by='null_counts'))
```

In future data processing we'll drop the columns with null values more than 50%

```python
# selecting the columns with more than 50% null data

train_drop = null_train[null_train['percentage'] > 50]
test_drop = null_test[null_test['percentage'] > 50]
```

### Plotting graphs

```python
train.drop(['Id'], axis = 1, inplace = True)
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(data = train.corr(numeric_only = True))
plt.show()
```

```python
# checking the relationship between numerical variables and target varaiable

num_cols = train.select_dtypes(include = 'number').columns
num_cols = num_cols.drop(['SalePrice'])
num_cols
```

```python
for i in num_cols:
    train.groupby(i)['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

```python


cat_cols = train.select_dtypes(include = 'object').columns
cat_cols
```

```python
for i in cat_cols:
    train.groupby(i)['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```
